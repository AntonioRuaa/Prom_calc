```python
# instalamos e importamos las librerías necesarias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as st
```

```python
# Creamos una lista vacía para los datos
L1 = []
L2 = []
L3 = []
L4 = []
L5 = []
```

```python
#Llenamos la lista con los dato del archivo
for line in open('20.dat'):
  columns = line.split()
  L1.append(columns)

for line in open('30.dat'):
  columns = line.split()
  L2.append(columns)

for line in open('50.dat'):
  columns = line.split()
  L3.append(columns)

for line in open('70.dat'):
  columns = line.split()
  L4.append(columns)

for line in open('90.dat'):
  columns = line.split()
  L5.append(columns)
```

```python
# Creamos un dataframe con los datos
df1 = pd.DataFrame(L1)
df1.rename(columns={0:'x', 1:'y', 2:'z', 3:'px', 4:'py', 5:'pz'})

df2 = pd.DataFrame(L2)
df2.rename(columns={0:'x', 1:'y', 2:'z', 3:'px', 4:'py', 5:'pz'})

df3 = pd.DataFrame(L3)
df3.rename(columns={0:'x', 1:'y', 2:'z', 3:'px', 4:'py', 5:'pz'})

df4 = pd.DataFrame(L4)
df4.rename(columns={0:'x', 1:'y', 2:'z', 3:'px', 4:'py', 5:'pz'})

df5 = pd.DataFrame(L5)
df5.rename(columns={0:'x', 1:'y', 2:'z', 3:'px', 4:'py', 5:'pz'})
```

## Datos de 20.dat

```python
#Calculamos los promedios para x
x1 = df1.iloc[:,0]
x1 = [float(i) for i in x1]
media1_1 = np.mean(x1)
des1_1 = np.std(x1)
skew1_1 = st.skew(x1)
kurt1_1 = st.kurtosis(x1)
print(media1_1, des1_1, skew1_1, kurt1_1)
```

```python
#Calculamos los promedios para px
x2 = df1.iloc[:,1]
x2 = [float(i) for i in x2]
media1_2 = np.mean(x2)
des1_2 = np.std(x2)
skew1_2 = st.skew(x2)
kurt1_2 = st.kurtosis(x2)
print(media1_2, des1_2, skew1_2, kurt1_2)
```

```python
# Creamos el histograma
y1 = [x1, x2]
y1_lab = ['x', 'px']
fig, ax = plt.subplots()
ax.hist(y1, bins = 15, label = y1_lab)
plt.axvline(media1_1, color = 'black', linestyle= '-', label = 'Media')
plt.axvline(media1_2, color = 'black', linestyle= '-')
plt.axvline(media1_1 + des1_1, color = 'red', linestyle= '--', label = 'Desviación')
plt.axvline(media1_1 - des1_1, color = 'red', linestyle= '--')
plt.axvline(media1_2 + des1_2, color = 'green', linestyle= '--')
plt.axvline(media1_2 - des1_2, color = 'green', linestyle= '--')
plt.text(1.01, 0.9, 'skewness x = '+str(round(skew1_1, 3)), transform = ax.transAxes)
plt.text(1.01, 0.85, 'skewness px = '+str(round(skew1_2, 3)), transform = ax.transAxes)
plt.text(1.01, 0.8, 'kurtosis x = '+str(round(kurt1_1, 3)), transform = ax.transAxes)
plt.text(1.01, 0.75, 'kurtosis px = '+str(round(kurt1_2, 3)), transform = ax.transAxes)
plt.title('Histograma de x y px en 20.dat')
plt.ylabel('Frecuencias')
plt.legend()
```
