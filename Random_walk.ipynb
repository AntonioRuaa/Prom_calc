```python
#instalamos las librerías necesarias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as rnd
```

```python
# Hacemos un for para el número de pasos
nt=200
dx = 1
X = np.zeros(nt)
Y = np.zeros(nt)


for i in range (nt):
    if i == 0:
        X[i]=0
        Y[i]=0
    else:
        XoY = rnd.randint(0, 1)
        if XoY == 0:
            sgn1 = rnd.randint(0, 1)
            if sgn1 == 0:
                X[i] = X[i-1]+dx
                Y[i] = Y[i-1]
            else:
                X[i] = X[i-1]-dx
                Y[i] = Y[i-1]
        else:
            sgn2 = rnd.randint(0, 1)
            if sgn2 == 0:
                Y[i] = Y[i-1]+dx
                X[i] = X[i-1]
            else:
                Y[i] = Y[i-1]-dx
                X[i] = X[i-1] 
```

```python
plt.plot(X,Y)
plt.title('Trayectoria del movimiento Browniano')
plt.ylabel('Movimiento en y')
plt.xlabel('Movimiento en X')
```

```python
n = range(nt)
D = np.zeros(nt)
for i in range (nt):
    D[i]=((X[i])**2+(Y[i])**2)**(0.5)
```

```python
plt.plot(n,D)
plt.title('Cuadrado de la distancia recorrida por número de tiro')
plt.ylabel('X2 + Y2')
plt.xlabel('n')
```
